A generic function to plot 2 variables in Zipf format within the same plot. We use the name of the raw input files as the labels.

```python
def plot_zipf(title, x_label, y_label, x_axis_code, y_axis_code, x_axis_lang, y_axis_lang, x_axis_lang2, y_axis_lang2, x_axis_lang3, y_axis_lang3, x_axis_lang4, y_axis_lang4, x_axis_lang5, y_axis_lang5, x_axis_lang6, y_axis_lang6, x_axis_lang7, y_axis_lang7, legend_labels):
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.plot(x_axis_code, y_axis_code, linewidth=2, color='r', label=legend_labels[0])
    plt.plot(x_axis_lang, y_axis_lang, linewidth=2, color='b', label=legend_labels[1])
    plt.plot(x_axis_lang2, y_axis_lang2, linewidth=2, color='g', label=legend_labels[2])
    plt.plot(x_axis_lang3, y_axis_lang3, linewidth=2, color='y', label=legend_labels[3])
    plt.plot(x_axis_lang4, y_axis_lang4, linewidth=2, color='c', label=legend_labels[4])
    plt.plot(x_axis_lang5, y_axis_lang5, linewidth=2, color='k', label=legend_labels[5])
    plt.plot(x_axis_lang6, y_axis_lang6, linewidth=2, color='m', label=legend_labels[6])
    plt.plot(x_axis_lang7, y_axis_lang7, linewidth=2, color='gray', label=legend_labels[7])
    plt.xscale("log")
    plt.yscale("log")
    plt.legend(loc="upper right")
    plt.show()
```

Sort the ngrams by their frequency, normalize the frequencies by the total token (or ngram) count, and plot on a log-scaled graph.

```python
def process_and_plot(processing_func: Callable, ngram_size: int, 
                    input_file_1: str, input_file_2: str, input_file_3: str,
                    input_file_4, input_file_5, input_file_6, input_file_7, input_file_8,
                    data_file_store_1: str, data_file_store_2: str, data_file_store_3: str, data_file_store_4,
                    data_file_store_5, data_file_store_6, data_file_store_7, data_file_store_8):
  
    dict_input, input_token_count = processing_func(input_file_1, ngram_size)
    dict_output, output_token_count = processing_func(input_file_2, ngram_size)
    dict_output2, output_token_count2 = processing_func(input_file_3, ngram_size)
    dict_output3, output_token_count3 = processing_func(input_file_4, ngram_size)
    dict_output4, output_token_count4 = processing_func(input_file_5, ngram_size)
    dict_output5, output_token_count5 = processing_func(input_file_6, ngram_size)
    dict_output6, output_token_count6 = processing_func(input_file_7, ngram_size)
    dict_output7, output_token_count7 = processing_func(input_file_8, ngram_size)
    
    """
    Here, we sort the ngrams by frequency, from largest to smallest.
    """
    sorted_list_input = sorted(dict_input.items(), key=lambda item: item[1], reverse=True)
    sorted_list_output = sorted(dict_output.items(), key=lambda item: item[1], reverse=True)
    sorted_list_output2 = sorted(dict_output2.items(), key=lambda item: item[1], reverse=True)
    sorted_list_output3 = sorted(dict_output3.items(), key=lambda item: item[1], reverse=True)
    sorted_list_output4 = sorted(dict_output4.items(), key=lambda item: item[1], reverse=True)
    sorted_list_output5 = sorted(dict_output5.items(), key=lambda item: item[1], reverse=True)
    sorted_list_output6 = sorted(dict_output6.items(), key=lambda item: item[1], reverse=True)
    sorted_list_output7 = sorted(dict_output7.items(), key=lambda item: item[1], reverse=True)

    """
    Create indices for the plot, and divide by token count (total number of ngrams in dataset)
    to get a true probability distribution.
    """
    x_axis_input = [x+1 for x in range(len(sorted_list_input))]
    y_axis_input = [[1.0 * item[1]/input_token_count] for item in sorted_list_input]

    x_axis_output = [x+1 for x in range(len(sorted_list_output))]
    y_axis_output = [[1.0 * item[1]/output_token_count] for item in sorted_list_output]
    
    x_axis_output2 = [x+1 for x in range(len(sorted_list_output2))]
    y_axis_output2 = [[1.0 * item[1]/output_token_count2] for item in sorted_list_output2]
    
    x_axis_output3 = [x+1 for x in range(len(sorted_list_output3))]
    y_axis_output3 = [[1.0 * item[1]/output_token_count3] for item in sorted_list_output3]
    
    x_axis_output4 = [x+1 for x in range(len(sorted_list_output4))]
    y_axis_output4 = [[1.0 * item[1]/output_token_count4] for item in sorted_list_output4]

    x_axis_output5 = [x+1 for x in range(len(sorted_list_output5))]
    y_axis_output5 = [[1.0 * item[1]/output_token_count5] for item in sorted_list_output5]
    
    x_axis_output6 = [x+1 for x in range(len(sorted_list_output6))]
    y_axis_output6 = [[1.0 * item[1]/output_token_count6] for item in sorted_list_output6]
    
    x_axis_output7 = [x+1 for x in range(len(sorted_list_output7))]
    y_axis_output7 = [[1.0 * item[1]/output_token_count7] for item in sorted_list_output7]

    legend_labels = ['CodeNN', 'NL', 'DeepCom1', 'DeepCom2', 'Docstring1', 'Docstring2', 'FunCom1', 'FunCom2']

    pickle.dump(y_axis_input, open(data_file_store_1, "wb"))
    pickle.dump(y_axis_output, open(data_file_store_2, "wb"))
    pickle.dump(y_axis_output2, open(data_file_store_3, "wb"))
    
    pickle.dump(y_axis_output3, open(data_file_store_4, "wb"))
    pickle.dump(y_axis_output4, open(data_file_store_5, "wb"))
    pickle.dump(y_axis_output5, open(data_file_store_6, "wb"))
    pickle.dump(y_axis_output6, open(data_file_store_7, "wb"))
    pickle.dump(y_axis_output7, open(data_file_store_8, "wb"))

    plot_zipf("Zipf Plot", "Log Freq Rank", "Log Freq", x_axis_input, y_axis_input, x_axis_output, y_axis_output, x_axis_output2, y_axis_output2, x_axis_output3, y_axis_output3, x_axis_output4, y_axis_output4, x_axis_output5, y_axis_output5, x_axis_output6, y_axis_output6, x_axis_output7, y_axis_output7, legend_labels)
    
```

These functions do the data collection for the 4 ngrams. The inner function parses the file (removing unnecesary tokens), and creates a dictionary where the key is the token and the associated value is the frequency of the token. We pass this data to `process_and_plot` for sorting and plotting.

```python
def plot_ngram_zipf(input_file_1: str, input_file_2: str, input_file_3: str, data_file_4, data_file_5, data_file_6, data_file_7,  data_file_8, data_file_store_1, data_file_store_2, data_file_store_3, data_file_store_4, data_file_store_5, data_file_store_6, data_file_store_7, data_file_store_8, ngram_size: int):
  
  def process_ngrams(input_file: str, ngram_size: int):
    ngram_count = 0
    n_prev_tokens = []
    ngram_dict = {}

    with open(input_file) as f:
      line = f.readline()
      while line:
        line = line.split(" ")
        prev_token = ""
        n_prev_tokens = []
        for token in line:
            if token == "\n":
                break
            
            n_prev_tokens.append(token)
            if len(n_prev_tokens) < ngram_size:
              continue
    
            ngram_count+=1
            cur_ngram = "-".join(n_prev_tokens)

            if cur_ngram not in ngram_dict:
                ngram_dict[cur_ngram] = 1
            else:
                ngram_dict[cur_ngram]+=1

            del n_prev_tokens[0]

        line = f.readline()

    return ngram_dict, ngram_count
  process_and_plot(process_ngrams, ngram_size, input_file_1, input_file_2, input_file_3, data_file_4, data_file_5, data_file_6, data_file_7,  data_file_8, data_file_store_1, data_file_store_2, data_file_store_3, data_file_store_4, data_file_store_5, data_file_store_6, data_file_store_7, data_file_store_8)
```

We have 5 changeable parameters here:

*   `ngram -> The specified ngram. Allowed values: "unigram", "bigram", "trigram", "tetragram"`
*   `data_file_1/data_file_2 -> Files where the data is located`
*    `data_file_store_1/data_file_store_2 -> Files you want to save frequencies in pickle format (must be .p files)`


NOTE: this function will calculate the zipf plots for two data files at a time, but will store pickle (.p) for you to accumulate all results into one graph.

```python
ngram_size = 1

# These data files correspond to the files you want to read from
data_file_1 = "../../data/plot-data/univariate/CodeNN_comments.txt" 
data_file_2 = "../../data/plot-data/univariate/NL_comments.txt"
data_file_3 = "../../data/plot-data/univariate/DeepCom1_comments.txt"

data_file_4 = "../../data/plot-data/univariate/DeepCom2_comments.txt"
data_file_5 = "../../data/plot-data/univariate/Docstring1_comments.txt"
data_file_6 = "../../data/plot-data/univariate/Docstring2_comments.txt"
data_file_7 = "../../data/plot-data/univariate/FunCom1_comments.txt"
data_file_8 = "../../data/plot-data/univariate/FunCom2_comments.txt"


# These files correspond to where you want to store the result pickle files
# so that you can plot everything together.
data_file_store_1 = "CodeNN_unigrams.p"
data_file_store_2 = "NL_unigrams.p"
data_file_store_3 = "DeepCom1_comments.p"

data_file_store_4 = "DeepCom2_comments.p"
data_file_store_5 = "Docstring1_comments.p"
data_file_store_6 = "Docstring2_comments.p"
data_file_store_7 = "FunCom1_comments.p"
data_file_store_8 = "FunCom2_comments.p"

plot_ngram_zipf(data_file_1, data_file_2, data_file_3, data_file_4, data_file_5, data_file_6, data_file_7,  data_file_8, data_file_store_1, data_file_store_2, data_file_store_3, data_file_store_4, data_file_store_5, data_file_store_6, data_file_store_7, data_file_store_8, ngram_size)
```
